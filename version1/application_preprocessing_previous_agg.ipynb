# solve null values on number

```python
total_rows = number_data.height

null_counts = number_data.null_count()

nulls_df_numbers = pl.DataFrame({
    "Column_Name": null_counts.columns,
    "Nulls_Count": null_counts.to_pandas().values.flatten()
})

nulls_df_numbers = nulls_df_numbers.with_columns(
    (pl.col("Nulls_Count") / total_rows * 100).alias("Percentage")
)

nulls_df_numbers = nulls_df_numbers.sort("Nulls_Count", descending=True)

nulls_df_numbers = nulls_df_numbers.filter(pl.col("Nulls_Count") > 0)
```

```python
nulls_df_numbers
```

```python
for col, percentage in zip(nulls_df_numbers['Column_Name'], nulls_df_numbers['Percentage']):
    if percentage <= 20:
        median_value = number_data.select(pl.col(col).median()).to_numpy()[0][0]  
        number_data = number_data.with_columns(
            pl.col(col).fill_null(median_value).alias(col)
        )
    elif percentage > 58:
        number_data = number_data.drop(col)

number_data_cleaned = number_data
number_data_cleaned
```

```python
total_rows = number_data_cleaned.height

null_counts = number_data_cleaned.null_count()

nulls_df_numbers = pl.DataFrame({
    "Column_Name": null_counts.columns,
    "Nulls_Count": null_counts.to_pandas().values.flatten()
})

nulls_df_numbers = nulls_df_numbers.with_columns(
    (pl.col("Nulls_Count") / total_rows * 100).alias("Percentage")
)

nulls_df_numbers = nulls_df_numbers.sort("Nulls_Count", descending=True)

nulls_df_numbers = nulls_df_numbers.filter(pl.col("Nulls_Count") > 0)
```

```python
previous_app_data=pd.read_csv("datasets\previous_application.csv")
```

```python
def agg(column_names, type_of_agg, id, data,table_name):
    aggregated_data = data.groupby(id).agg({col: type_of_agg for col in column_names}).reset_index()
    new_column_names = {col: f"{col}_{type_of_agg}_{table_name}" for col in column_names}
    aggregated_data.rename(columns=new_column_names, inplace=True)

    return aggregated_data
```

```python

aggs_max_columns = [
    "DAYS_DECISION",
    "CNT_PAYMENT",
    "DAYS_FIRST_DRAWING",
    "DAYS_FIRST_DUE",
    "DAYS_LAST_DUE_1ST_VERSION",
    "DAYS_LAST_DUE",
    "DAYS_TERMINATION",
    "NFLAG_INSURED_ON_APPROVAL",
]
agg_max_for_previous_columns=agg(aggs_max_columns,"mean","SK_ID_CURR",previous_app_data,"previous_app_data")
merge_max_previous_app=new_app_data.merge(agg_max_for_previous_columns,how='left',on='SK_ID_CURR')
```

```python

columns = ["DAYS_BIRTH_YEARS", "DAYS_REGISTRATION_Year", "DAYS_ID_PUBLISH_Years", "DAYS_EMPLOYED_Years"]

# Function to calculate IQR and replace outliers with NaN
def replace_outliers_with_nan(df, col):
    q1 = np.percentile(df[col], 25, interpolation="midpoint")
    q3 = np.percentile(df[col], 75, interpolation="midpoint")
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    df[col] = df[col].apply(lambda x: np.nan if x < lower_bound or x > upper_bound else x)

for col in columns:
    replace_outliers_with_nan(new_app_data, col)
```
