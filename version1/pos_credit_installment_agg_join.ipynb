# agg function

```python
def agg(column_names, type_of_agg, id, data,table_name):
    aggregated_data = data.groupby(id).agg({col: type_of_agg for col in column_names}).reset_index()
    new_column_names = {col: f"{col}_{type_of_agg}_{table_name}" for col in column_names}
    aggregated_data.rename(columns=new_column_names, inplace=True)

    return aggregated_data
```
